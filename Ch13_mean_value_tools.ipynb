```python
# 1. Import necessary modules

import sys
sys.path.append('utils')
from mean_values_tools import *
import sympy as sp

x = sp.symbols('x', real=True)
```

```python
# Rolle's Theorem Example
# Solved 1
f = x**2 - 4*x + 3
rolles = rolles_theorem(f, a=1, b=3)
print("Rolle's Theorem critical points:", rolles)
```

```python
# Mean Value Theorem Example
# Solved 3
f = 3*x**2 + 4*x - 3
slope, c_vals = mean_value_theorem(f, a=1, b=3)
slope, c_vals
print("Mean Value Theorem slope:", slope)
print("Mean Value Theorem critical points:", c_vals)
```

```python
# Cauchy MVT Example
#Solved 4
f = 3*x +2
g = x**2 + 1
ratio, cvals = cauchy_mvt(f, g, a=1, b=4)
ratio, cvals
print("Cauchy MVT ratio:", ratio)
print("Cauchy MVT critical points:", cvals)
```

```python
# Higher MVT Example
# f(a)=f(b)=0 implies f''(c)=0 for some c

f = x**4 - 5*x**2
#higher_mvt(f, a=-2, b=2, n=2)
print("Higher MVT critical points:", higher_mvt(f, a=-2, b=2, n=2))

```

```python
# Increasing / Decreasing Intervals
# Solved 10
f = x**5 + 20*x -6
crit, intervals = monotonic_intervals(f)
crit, intervals
print("Critical points:", crit)
print("Monotonic intervals:", intervals)
```
