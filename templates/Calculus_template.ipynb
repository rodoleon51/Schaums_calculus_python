```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# ---- SYMBOLIC SETUP ----
x = sp.symbols('x')
```

```python
# Example function
f = sp.sin(x) * sp.exp(-x)
```

```python
# Derivative
f_prime = sp.diff(f, x)
```

```python
# Integral
f_int = sp.integrate(f, x)
```

```python
print("f(x)      =", f)
print("f'(x)     =", f_prime)
print("∫f(x) dx  =", f_int)
```

```python
# ---- NUMERIC EVALUATION ----
f_numeric = sp.lambdify(x, f, 'numpy')

x_vals = np.linspace(0, 10, 400)
y_vals = f_numeric(x_vals)
```

```python
# ---- PLOT ----
plt.figure(figsize=(8, 4))
plt.plot(x_vals, y_vals)
plt.title("f(x) = sin(x) * e^{-x}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

```python
# ---- EXTRA: Evaluate a limit & Taylor series ----
limit_at_inf = sp.limit(f, x, sp.oo)
taylor_expansion = sp.series(f, x, 0, 6)

print("Limit as x→∞:", limit_at_inf)
print("Taylor series at x=0:", taylor_expansion)
```
