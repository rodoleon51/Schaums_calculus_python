##### Chapter 11 - Implicit Differentiation

```python
# Solved problems
# 1. Implicit Differentiation
# (x**2)*y - x*y**2 + x**2 + y**2 = 0

from sympy import symbols, diff, simplify, pprint

# Define variables
x, y = symbols('x y')

# Given implicit equation: F(x, y) = 0
expr = x**2 * y - x * y**2 + x**2 + y**2

# Step 1: Compute partial derivatives
F_x = diff(expr, x)   # ∂F/∂x
F_y = diff(expr, y)   # ∂F/∂y

# Step 2: First derivative: dy/dx = -F_x / F_y
dy_dx = simplify(-F_x / F_y)

print("dy/dx:")
dy_dx
```

```python
# 2. Second Derivative
# Given implicit equation: x**2 - x*y + y**2 - 3 = 0

from sympy import symbols, diff, simplify, pprint

# Define variables
x, y = symbols('x y')

# Given implicit equation: F(x, y) = 0
expr = x**2 - x*y + y**2 - 3

# Step 1: Compute partial derivatives
F_x = diff(expr, x)   # ∂F/∂x
print("F_x:", F_x)
F_y = diff(expr, y)   # ∂F/∂y
print("F_y:", F_y)

# Step 2: First derivative: dy/dx = -F_x / F_y
dy_dx = simplify(-F_x / F_y)

# print("dy/dx:")
dy_dx

# Step 3: Second derivative using the formula for implicit differentiation
# d²y/dx² = - (∂²F/∂x² + 2(∂²F/∂x∂y)(dy/dx) + ∂²F/∂y²(dy/dx)²) / (∂F/∂y)

F_xx = diff(F_x, x)        # ∂²F/∂x²
F_xy = diff(F_x, y)        # ∂²F/∂x∂y
F_yy = diff(F_y, y)        # ∂²F/∂y²

numerator = F_xx + 2*F_xy*dy_dx + F_yy*(dy_dx)**2
d2y_dx2 = simplify(-numerator / F_y)

print("\nd²y/dx²:")
d2y_dx2
```

```python
# ------------------------------------------------------------
#  SymPy – first and second derivatives of the implicit curve
#  f(x,y) = x² - x y + y² - 3 = 0
# ------------------------------------------------------------

import sympy as sp

# ------------------------------------------------------------------
# 1. Define symbols and the implicit function
# ------------------------------------------------------------------
x, y = sp.symbols('x y', real=True)
f = x**2 - x*y + y**2 - 3

print("Implicit equation:")
sp.pprint(f)
print("\n" + "="*60 + "\n")

# ------------------------------------------------------------------
# 2. First partial derivatives
# ------------------------------------------------------------------
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)

print("First partial derivatives:")
print("∂f/∂x =", end=" ")
sp.pprint(df_dx)
print("∂f/∂y =", end=" ")
sp.pprint(df_dy)
print("\n" + "="*60 + "\n")

# ------------------------------------------------------------------
# 3. Second partial derivatives (Hessian)
# ------------------------------------------------------------------
d2f_dx2 = sp.diff(df_dx, x)      # f_xx
d2f_dxdy = sp.diff(df_dx, y)     # f_xy  (same as f_yx)
d2f_dy2 = sp.diff(df_dy, y)      # f_yy

print("Second partial derivatives:")
print("∂²f/∂x²  =", end=" ")
sp.pprint(d2f_dx2)
print("∂²f/∂x∂y =", end=" ")
sp.pprint(d2f_dxdy)
print("∂²f/∂y²  =", end=" ")
sp.pprint(d2f_dy2)
print("\n" + "="*60 + "\n")

# ------------------------------------------------------------------
# 4. (Optional) Evaluate at a point on the curve
# ------------------------------------------------------------------
# Find a numeric point that satisfies f(x,y)=0, e.g. (x=1, y=1)
point = {x: 1, y: 1}
if f.subs(point) == 0:
    print(f"Evaluating at the point ({point[x]}, {point[y]}):")
    print("  f      =", f.subs(point))
    print("  ∂f/∂x  =", df_dx.subs(point))
    print("  ∂f/∂y  =", df_dy.subs(point))
    print("  ∂²f/∂x² =", d2f_dx2.subs(point))
    print("  ∂²f/∂x∂y=", d2f_dxdy.subs(point))
    print("  ∂²f/∂y² =", d2f_dy2.subs(point))
else:
    print("Chosen point does not lie on the curve – skipping numeric evaluation.")

# ------------------------------------------------------------------
# 5. (Optional) Lambdify for fast numeric evaluation
# ------------------------------------------------------------------
# Convert symbolic expressions to NumPy-callable functions
df_dx_num = sp.lambdify((x, y), df_dx, 'numpy')
df_dy_num = sp.lambdify((x, y), df_dy, 'numpy')
d2f_dx2_num = sp.lambdify((x, y), d2f_dx2, 'numpy')
d2f_dxdy_num = sp.lambdify((x, y), d2f_dxdy, 'numpy')
d2f_dy2_num = sp.lambdify((x, y), d2f_dy2, 'numpy')

# Example usage with NumPy arrays
import numpy as np
X = np.array([0.0, 1.0, 2.0])
Y = np.array([3.0, 1.0, 0.0])      # note: not necessarily on the curve
print("\nNumeric evaluation at points (X,Y):")
print("∂f/∂x  =", df_dx_num(X, Y))
print("∂f/∂y  =", df_dy_num(X, Y))
print("∂²f/∂x² =", d2f_dx2_num(X, Y))
print("∂²f/∂x∂y=", d2f_dxdy_num(X, Y))
print("∂²f/∂y² =", d2f_dy2_num(X, Y))
```
