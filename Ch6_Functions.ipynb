## Chapter 5 - Functions

```python
import sympy as sp
from sympy import symbols, Eq, solve, lambdify
sp.init_printing()
```

```python
# 9  Evaluate the function f(x) = x^2 - 2x + 3 at x = 3, -3, 5, and -5. 

# Define the symbol and function
x = symbols('x')
f = x**2 - 2*x + 3

# Create a numerical function
f_num = lambdify(x, f)

# List of numbers to evaluate
numbers = [3, -3, 5, -5]

# Evaluate and print results
results = {n: f_num(n) for n in numbers}
results
```

```python
# 11  Find f(x) for x^2 - 54x + 6 = 0.

# Define the symbol and function
x = symbols('x')
f = x**2 - 54*x + 6

# Create a numerical function
f_num = lambdify(x, f)

# List of numbers to evaluate
numbers = [0, 3, -2]

# Evaluate and print results
results = {n: f_num(n) for n in numbers}
results
```
