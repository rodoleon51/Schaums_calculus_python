```python
from sympy import symbols, plot, simplify

# Define symbols
x, y = symbols('x y')

# Define the expression
expr = x**2 + y**2 + 4*y - 4

# Substitute y = x^2
expr_sub = expr.subs(y, x**2)

# Simplify the expression
simplified_expr = simplify(expr_sub)

# Plot the simplified expression
plot(simplified_expr, (x, -2, 2), title='Plot of x^2 + y^2 + 4y - 4 where y = x^2')
```

## Ellipses

```python
# Plot Ellipse: x²/a² + y²/b² = 1
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the ellipse: x²/a² + y²/b² = 1
a = 3.0
b = 2.0

# Generate x values from -3 to 3 with step 0.5
x_values = np.arange(-3, 3.25, 0.25)  # 3.5 to include 3.0

# Calculate corresponding y values (upper and lower parts)
# y = ±b * sqrt(1 - x²/a²)
valid_x = np.abs(x_values) <= a  # Only where |x| <= a
y_upper = np.zeros_like(x_values)
y_lower = np.zeros_like(x_values)

y_upper[valid_x] = b * np.sqrt(1 - (x_values[valid_x]/a)**2)
y_lower[valid_x] = -b * np.sqrt(1 - (x_values[valid_x]/a)**2)

# For clean printing, we'll show both branches as separate points
# But for plotting, we'll combine them
x_plot = np.concatenate([x_values[valid_x], x_values[valid_x][::-1]])  # full closed loop
y_plot = np.concatenate([y_upper[valid_x], y_lower[valid_x][::-1]])

# Print x and y values (only valid points, upper then lower)
print("x values (valid):", [round(float(x), 3) for x in x_values[valid_x]])
print("y values (upper):", [round(float(y), 3) for y in y_upper[valid_x]])
print("y values (lower):", [round(float(y), 3) for y in y_lower[valid_x]])

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_plot, 'bo-', linewidth=2, markersize=6, label=f'Ellipse: x²/{a}² + y²/{b}² = 1')
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Ellipse: $\\frac{{x^2}}{{{a}^2}} + \\frac{{y^2}}{{{b}^2}} = 1$')
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.axis('equal')  # Important! Ensures ellipse isn't stretched
plt.legend()

# Save the plot BEFORE plt.show()
save_path = 'C://Users//rodol//Documents//Workspace_Python//Calculus//Schaums//my_plots//'

# Method 1: Save as PNG (default)
plt.savefig(save_path + 'ellipse_plot_2.png')

# Method 2: Save as PDF
plt.savefig(save_path + 'ellipse_plot_ellipse.pdf')

# Method 3: Save with high DPI for better quality
plt.savefig(save_path + 'ellipse_plot_hd_ellipse.png', dpi=300, bbox_inches='tight')

# Method 4: Save with transparent background
plt.savefig(save_path + 'ellipse_plot_transparent_ellipse.png', transparent=True, bbox_inches='tight')

print("Plots saved as:")
print("- ellipse_plot.png")
print("- ellipse_plot.pdf") 
print("- ellipse_plot_hd.png (high quality)")
print("- ellipse_plot_transparent.png (transparent background)")

#plt.show()
```

## Hyperbolas

```python
# Plot Hyperbola: x²/a² - y²/b² = 1 
import matplotlib.pyplot as plt
%matplotlib widget

# Parameters
a = 3.0
b = 2.0

# Data (same as before)
x_right = np.linspace(a, 10, 200)
y_upper_right = b * np.sqrt((x_right/a)**2 - 1)
y_lower_right = -y_upper_right

x_left = np.linspace(-10, -a, 200)
y_upper_left = b * np.sqrt((x_left/a)**2 - 1)
y_lower_left = -y_upper_left

# Discrete points
x_discrete = np.arange(-6, 6.5, 0.5)
valid = np.abs(x_discrete) >= a
x_valid = x_discrete[valid]
y_upper_disc = b * np.sqrt((x_valid/a)**2 - 1)
y_lower_disc = -y_upper_disc
sort_idx = np.argsort(x_valid)
x_valid = x_valid[sort_idx]
y_upper_disc = y_upper_disc[sort_idx]
y_lower_disc = y_lower_disc[sort_idx]


# Print x and y values before plotting
print("HYPERBOLA PLOTTED VALUES:")
print("="*50)
print(f"Parameters: a={a}, b={b}")
print()

print("Right branch (x ≥ a):")
print("First 5 points:")
for i in range(min(5, len(x_right))):
    print(f"  x = {x_right[i]:6.2f}, y_upper = {y_upper_right[i]:6.2f}, y_lower = {y_lower_right[i]:6.2f}")
if len(x_right) > 5:
    print(f"  ... ({len(x_right)-10} points in middle) ...")
    print("Last 5 points:")
    for i in range(max(0, len(x_right)-5), len(x_right)):
        print(f"  x = {x_right[i]:6.2f}, y_upper = {y_upper_right[i]:6.2f}, y_lower = {y_lower_right[i]:6.2f}")

print()
print("Left branch (x ≤ -a):")
print("First 5 points:")
for i in range(min(5, len(x_left))):
    print(f"  x = {x_left[i]:6.2f}, y_upper = {y_upper_left[i]:6.2f}, y_lower = {y_lower_left[i]:6.2f}")
if len(x_left) > 5:
    print(f"  ... ({len(x_left)-10} points in middle) ...")
    print("Last 5 points:")
    for i in range(max(0, len(x_left)-5), len(x_left)):
        print(f"  x = {x_left[i]:6.2f}, y_upper = {y_upper_left[i]:6.2f}, y_lower = {y_lower_left[i]:6.2f}")

print()
print("Discrete sample points:")
for i in range(len(x_valid)):
    print(f"  x = {x_valid[i]:6.2f}, y_upper = {y_upper_disc[i]:6.2f}, y_lower = {y_lower_disc[i]:6.2f}")

print("="*50)
print()


# === CRITICAL: SQUARE FIGURE ===
fig = plt.figure(figsize=(6, 6))  # SQUARE! This is the key

# === PLOT ===
plt.plot(x_right, y_upper_right, 'b-', linewidth=2)
plt.plot(x_right, y_lower_right, 'b-', linewidth=2)
plt.plot(x_left,  y_upper_left,  'b-', linewidth=2, label='Hyperbola')
plt.plot(x_left,  y_lower_left,  'b-', linewidth=2)

plt.plot(x_valid, y_upper_disc, 'ro', markersize=5, alpha=0.7, label='Sample points')
plt.plot(x_valid, y_lower_disc, 'ro', markersize=5, alpha=0.7)

# Asymptotes
x_asym = np.linspace(-10, 10, 100)
plt.plot(x_asym,  (b/a)*x_asym, 'k--', alpha=0.5, label='Asymptotes')
plt.plot(x_asym, -(b/a)*x_asym, 'k--', alpha=0.5)

# Vertices
plt.plot([a, -a], [0, 0], 'go', markersize=10, label='Vertices')
plt.text(a + 0.3, 0.5, f'({a}, 0)', fontsize=10, color='green')
plt.text(-a - 1.8, 0.5, f'({-a}, 0)', fontsize=10, color='green')

# Formatting
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Hyperbola: $\\frac{{x^2}}{{{a}^2}} - \\frac{{y^2}}{{{b}^2}} = 1$')
plt.axhline(0, color='k', alpha=0.3)
plt.axvline(0, color='k', alpha=0.3)

# === EQUAL ASPECT + FULL DATA RANGE ===
plt.axis('equal')
plt.xlim(-10.5, 10.5)
plt.ylim(-10.5, 10.5)

# === FORCE WIDGET TO RESIZE ===
fig.canvas.draw_idle()
plt.tight_layout()

# === LEGEND ===
plt.legend(loc='upper left')

# === SAVE (perfect, with equal aspect) ===
save_path = 'C://Users//rodol//Documents//Workspace_Python//Calculus//Schaums//my_plots//'
plt.savefig(save_path + 'hyperbola_plot_fixed.png', dpi=150, bbox_inches='tight')
plt.savefig(save_path + 'hyperbola_plot_fixed.pdf', bbox_inches='tight')
plt.savefig(save_path + 'hyperbola_plot_hd_fixed.png', dpi=300, bbox_inches='tight')
plt.savefig(save_path + 'hyperbola_plot_transparent_fixed.png', transparent=True, bbox_inches='tight')

plt.show()
```

## Plot Power  Functions: y = kx^n 
### (Generalized Conic-Like Curves)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget

# ------------------------------------------------------------------
# 1. USER INPUT
# ------------------------------------------------------------------
print("Enter the parameters for plotting y = kx^n")
k = float(input("Enter coefficient k: "))
x_min = float(input("Enter minimum x value: "))
x_max = float(input("Enter maximum x value: "))
n = float(input("Enter exponent n: "))

if x_min >= x_max:
    print("Error: x_min must be less than x_max")
    raise SystemExit

# ------------------------------------------------------------------
# 2. DATA
# ------------------------------------------------------------------
x = np.linspace(x_min, x_max, 1000)
exponents = [n]
colors    = ['red']
labels    = [f'y = {k}x^{n}']

# ------------------------------------------------------------------
# 3. FIGURE – *tight layout + explicit padding*
# -------------------------------------------------------------------
plt.close('all')                                 # start fresh
fig = plt.figure(figsize=(6, 4))                 # a little smaller than before
fig.tight_layout(pad=4.0)                        # <-- forces everything inside

# ------------------------------------------------------------------
# 4. PLOT
# ------------------------------------------------------------------
for n_val, col, lab in zip(exponents, colors, labels):
    if n_val % 1 != 0:                 # fractional power → only positive x
        x_plot = x[x > 0]
        if len(x_plot):
            y = k * (x_plot ** n_val)
            plt.plot(x_plot, y, color=col, linewidth=2, label=lab)
    else:
        y = k * (x ** n_val)
        plt.plot(x, y, color=col, linewidth=2, label=lab)

# ------------------------------------------------------------------
# 5. DECORATIONS
# ------------------------------------------------------------------
plt.axhline(0, color='black', ls='-', alpha=0.3)
plt.axvline(0, color='black', ls='-', alpha=0.3)
plt.grid(True, alpha=0.3)

plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Power Function: y = {k}x^{n}')
plt.xlim(x_min, x_max)

# ------------------------------------------------------------------
# PRINT PLOTTED (x, y) VALUES
# ------------------------------------------------------------------
print("\n" + "="*60)
print("PLOTTED POINTS (x, y):")
print("="*60)

for n_val, color, label in zip(exponents, colors, labels):
    print(f"\nCurve: {label}")
    if n_val % 1 != 0:  # fractional exponent → only positive x
        x_plot = x[x > 0]
        if len(x_plot) == 0:
            print("  No positive x values in range → nothing plotted.")
            continue
        y_plot = k * (x_plot ** n_val)
    else:
        x_plot = x
        y_plot = k * (x ** n_val)

    # Print first 5, last 5, and indicate if there are more
    n_points = len(x_plot)
    if n_points <= 10:
        for xi, yi in zip(x_plot, y_plot):
            print(f"  x = {xi:8.4f},  y = {yi:12.6f}")
    else:
        print("  First 5 points:")
        for xi, yi in zip(x_plot[:5], y_plot[:5]):
            print(f"    x = {xi:8.4f},  y = {yi:12.6f}")
        print(f"  ... ({n_points - 10} points skipped) ...")
        print("  Last 5 points:")
        for xi, yi in zip(x_plot[-5:], y_plot[-5:]):
            print(f"    x = {xi:8.4f},  y = {yi:12.6f}")

print("="*60 + "\n")


# ------------------------------------------------------------------
# 6. Y-LIMITS (same logic you already had)
# ------------------------------------------------------------------
all_y = []
for n_val in exponents:
    if n_val % 1 != 0:
        xp = x[x > 0]
        if len(xp):
            all_y.extend(k * (xp ** n_val))
    else:
        all_y.extend(k * (x ** n_val))

if all_y:
    y_min, y_max = min(all_y), max(all_y)
    y_range = y_max - y_min or 1
    plt.ylim(y_min - 0.1*y_range, y_max + 0.1*y_range)

# ------------------------------------------------------------------
# 7. LEGEND – placed *outside* the axes so it never clips the curve
# ------------------------------------------------------------------
plt.legend(loc='upper left', bbox_to_anchor=(1.01, 1), borderaxespad=0)

# ------------------------------------------------------------------
# 8. FORCE A FULL RE-LAYOUT BEFORE SHOWING
# ------------------------------------------------------------------
fig.canvas.draw_idle()          # recompute layout
plt.pause(0.01)                 # tiny pause lets the GUI update

# ------------------------------------------------------------------
# 9. SAVE 
# ------------------------------------------------------------------
save_path = 'C://Users//rodol//Documents//Workspace_Python//Calculus//Schaums//my_plots//'
for name, kw in [
    ('power_functions_plot.png',               dict(dpi=150)),
    ('power_functions_plot.pdf',               dict()),
    ('power_functions_hd_plot.png',            dict(dpi=300)),
    ('power_functions_transparent_plot.png',   dict(transparent=True)),
]:
    plt.savefig(save_path + name, bbox_inches='tight', **kw)

print("Plots saved:")
print("- power_functions_plot.png")
print("- power_functions_plot.pdf")
print("- power_functions_hd_plot.png (high quality)")
print("- power_functions_transparent_plot.png (transparent)")

# ------------------------------------------------------------------
# 10. SHOW 
# ------------------------------------------------------------------
# plt.show()
```
