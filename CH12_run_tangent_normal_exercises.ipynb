```python
# ============================================
# run_exercises.py
# Demonstrates usage of tangent_normal_tools.py
# ============================================

import sys
sys.path.append('utils')
import sympy as sp

from sympy import pprint
from tangent_normal_tools import (
    tangent_normal_explicit,
    tangent_to_implicit_with_slope,
    horizontal_vertical_tangents,
    plot_explicit_tangent_normal,
    plot_implicit_tangent
)

#from utils.tangent_normal_tools import *

x, y = sp.symbols('x y', real=True)
```

```python
# -------------------------------------------------------------
# EXAMPLE 1 â€” explicit function
# Tangent and normal to y = x^3 - 2x^2 + 4 at x=2
# -------------------------------------------------------------
fx = x**3 - 2*x**2 + 4
print("=== Example 1 ===")
res1 = tangent_normal_explicit(fx, x0=2)

print("y(2) =", res1["y0"])
print("Slope f'(2) =", res1["slope"])
print("Tangent line:", res1["tangent"])
print("Normal line:", res1["normal"])
print()
```

```python
fx = x**3 - 2*x**2 + 4 
plot_explicit_tangent_normal(fx, x0=2, x_range=(0,4))
```

```python
# -------------------------------------------------------------
# EXAMPLE 2 â€” tangent line with GIVEN slope to an ellipse
# Find tangent with slope -2/9 to 4x^2 + 9y^2 = 40
# -------------------------------------------------------------
F2 = 4*x**2 + 9*y**2 - 40
print("=== Example 2 ===")
res2 = tangent_to_implicit_with_slope(F2, m=-sp.Rational(2, 9))
print("Points where slope = -2/9:")
print(res2)
print()
```

```python
# def plot_implicit_tangent(F, point, x_range=(-10, 10), y_range=(-10, 10), num=400):
point = -1, -2
point = 1, 2
plot_implicit_tangent(F2, point, x_range=(-6, 6), y_range=(-4, 4), num=400)
```

```python
# -------------------------------------------------------------
# EXAMPLE 3 â€” horizontal & vertical tangents
# Curve: x^2 - x*y + y^2 = 27
# -------------------------------------------------------------
F3 = x**2 - x*y + y**2 - 27
print("=== Example 3 ===")
horiz, vert = horizontal_vertical_tangents(F3)

print("Horizontal tangent points:")
print(horiz)

print("\nVertical tangent points:")
print(vert)
```
